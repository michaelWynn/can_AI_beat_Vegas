```python
# Import modified data
import pandas as pd
import numpy as np
df = pd.read_csv('preppedData/games_advanced.csv')

# Drop unnecessary identifiers and regression target
df = df.drop(columns=['ID','GAME_DATE_EST','GAME_ID','SEASON','HOME_TEAM_ID','AWAY_TEAM_ID','MARGIN'])
df.info()
```

```python
# Import packages for data processing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score

# Import packages for neural network
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Seperate columns
x = df.values[:, 0:((len(df.columns))-1)]
y = df.values[:, ((len(df.columns))-1)]

# Feature scaling
scaler = MinMaxScaler(feature_range=(0, 1))
x_scaled = scaler.fit_transform(x)

# Train(70)/Test(15)/Val(15) Split
x_train, x_test_and_val, y_train, y_test_and_val = train_test_split(x_scaled, y, test_size = 0.3, random_state = 100)
x_test, x_val, y_test, y_val = train_test_split(x_test_and_val, y_test_and_val, test_size = 0.5, random_state = 100)
```

## Creation and training of neural network 1

```python
model = Sequential([Dense(44, activation='relu', input_shape=(((len(df.columns))-1),)), 
                    Dense(22, activation='relu'), 
                    Dense(1, activation='sigmoid'),])
model.compile(optimizer='sgd',loss='binary_crossentropy', metrics=['accuracy'])
hist = model.fit(x_train, y_train, batch_size=32, epochs=500, validation_data=(x_val, y_val))
```

## Calculating metrics for neural network 1

```python
# predict probabilities for test set
y_pred = model.predict(x_test, verbose=1)
# predict classes for test set
y_pred_classes = (model.predict(x_test) > 0.5).astype("int32")
# reduce to 1d array
y_pred = y_pred[:, 0]
y_pred_classes = y_pred_classes[:, 0]

# confusion matrix
print(confusion_matrix(y_test, y_pred_classes))

print()
# accuracy: (tp + tn) / (p + n)
print('Accuracy: %f' % accuracy_score(y_test, y_pred_classes))

# precision tp / (tp + fp)
print('Precision: %f' % precision_score(y_test, y_pred_classes))

# recall: tp / (tp + fn)
print('Recall: %f' % recall_score(y_test, y_pred_classes))
```

## Graphical analysis of neural network 1

```python
import matplotlib.pyplot as plt
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Val'], loc='upper right')
plt.show()
```

```python
plt.plot(hist.history['accuracy'])
plt.plot(hist.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Val'], loc='lower right')
plt.show()
```

## Creation and training of neural network 2
### Added regularization and dropout

```python
from keras.layers import Dropout
from keras import regularizers

model_2 = Sequential([Dense(64, activation='relu', kernel_regularizer=regularizers.l2(0.01), input_shape=(((len(df.columns))-1),)), Dropout(0.3), 
                      Dense(256, activation='relu', kernel_regularizer=regularizers.l2(0.01)), Dropout(0.3),  
                      Dense(256, activation='relu', kernel_regularizer=regularizers.l2(0.01)), Dropout(0.3), 
                      Dense(1, activation='sigmoid', kernel_regularizer=regularizers.l2(0.01)),])
model_2.compile(optimizer='sgd',loss='binary_crossentropy',metrics=['accuracy'])
hist_2 = model_2.fit(x_train, y_train, batch_size=32, epochs=100, validation_data=(x_val, y_val))
```

## Calculating metrics for neural network 2

```python
# predict probabilities for test set
y_pred = model_2.predict(x_test, verbose=1)
# predict classes for test set
y_pred_classes = (model_2.predict(x_test) > 0.5).astype("int32")
# reduce to 1d array
y_pred = y_pred[:, 0]
y_pred_classes = y_pred_classes[:, 0]

# confusion matrix
print(confusion_matrix(y_test, y_pred_classes))

# accuracy: (tp + tn) / (p + n)
print('Accuracy: %f' % accuracy_score(y_test, y_pred_classes))

# precision tp / (tp + fp)
print('Precision: %f' % precision_score(y_test, y_pred_classes))

# recall: tp / (tp + fn)
print('Recall: %f' % recall_score(y_test, y_pred_classes))
```

## Graphical analysis of neural network 2

```python
plt.plot(hist_2.history['loss'])
plt.plot(hist_2.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Val'], loc='upper right')
plt.show()
```

```python
plt.plot(hist_2.history['accuracy'])
plt.plot(hist_2.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Val'], loc='lower right')
plt.show()
```
